# Product Name Generator
Multi-class classification

```python
import os
import pandas as pd
import tensorflow as tf
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Embedding, Dropout
```

```python
# Ensuring GPU is available
gpu_info = tf.config.list_physical_devices("GPU")

# enable GPU
os.environ["TF_FORCE_GPU_ALLOW_GROWTH"]="true"

if (len(gpu_info) > 0):
    tf.config.experimental.set_memory_growth(gpu_info[0], enable=True)

print(f'GPU info: {gpu_info}')
print(f'Number of GPUs: {len(gpu_info)}')
print(f'Tensorflow version: {tf.__version__}')
print(f'Keras version: {tf.keras.__version__}')
```

# Dataset (quick) analysis

```python
df = pd.read_excel('./data-input/small_sentences_dataset_FR_clean.xlsx', 
                   # reading xlsx files
                   engine='openpyxl',
                   header=None, 
                   names=['product_name'])

df.head()
```

```python
# Confirming it's only one column
print(df.shape)
print('\n')
print(df.describe())

# confirmation of an example of a duplicate
assert len(df.loc[df['product_name'] == 'Valeur refuge']) > 1
assert len(df.loc[df['product_name'] == 'Maoris']) == 17
```

From the difference between count and unique, we can see that there are quite a few duplicates. For example, the product name 'Maoris' is repeated 17 times. The neural network won't need duplicates on the inputs, so they're removed here. 

But before removing all the duplicates, we ensure that the text is on the same level. Meaning, that all the product names are lower cased.


# Text preprocessing

```python
def preprocess_text(text):
    # lowercase & strip spaces in beginning and end
    # keras' tokenizer can also do this.
    text = str(text).lower().strip()
    
    return text
```

```python
df_processed = pd.DataFrame(df['product_name'].apply(preprocess_text))

print(df_processed.describe())
```

```python
# let's remove duplicates
df_nodup = df_processed.drop_duplicates()

# ensure that we have no more duplicates
assert len(df_nodup.loc[df_nodup['product_name'] == 'maoris']) == 1

df_nodup.describe()
```

```python
corpus = df_nodup['product_name'].to_list()
print(f'Size of corpus: {len(corpus)}')
print(f'Corpus preview: {corpus[:5]}')
```

# Tokenization

```python
# determine number of unique tokens (input_dim)

tokenizer = Tokenizer()
tokenizer.fit_on_texts(corpus)

#input_dim
vocab_size = len(tokenizer.word_index) + 1

print(f'total number of diferent words: {vocab_size - 1}')
```

# Sequencing 
Using n-grams of words.

```python
# sequences

input_sequences = []
train_length = 1

for product_name in corpus:
    # transforming each product name into a sequence of integers. 
    token_list = tokenizer.texts_to_sequences([product_name])[0]
    
    # creating an n-gram sequence for each of the product names.
    for i in range(train_length, len(token_list)):
        input_sequences.append(token_list[:i+1])

# making sure the output is the desired
assert len(input_sequences) > vocab_size
# and preview it
input_sequences[:10]

# for item in input_sequences[:10]:
#     for index in item:
#         print(tokenizer.index_word[index])
        
#     print('\n')
```

```python
# now for some padding, so that all the sequences have the same length.
max_sequence_len = max([len(x) for x in input_sequences])
assert max_sequence_len > 1

input_sequences_padded = np.array(pad_sequences(input_sequences, maxlen=max_sequence_len, padding='pre'))
```

# Preparing predictor and label

```python
X = input_sequences_padded[:,:-1]

# one-hot encoding
y = to_categorical(input_sequences_padded[:,-1], num_classes=vocab_size)
```

# Create model

```python
print(X.shape, y.shape, vocab_size)
```

```python
# from tensorflow.keras.losses import sparse_categorical_crossentropy
def create_model(max_sequence_len, vocab_size, X):    
    model = Sequential()
    
    input_len = X.shape[1]
    
    # Input layer
    model.add(Embedding(
        input_dim=vocab_size,
        
        # To avoid long product names
        output_dim=input_len,
        
        input_length=input_len))
    
    # Hidden layer(s)
#     model.add(LSTM(units=50, return_sequences=True))
    
    model.add(LSTM(units=50))
    
#     model.add(Dense(units=50, activation='relu'))
    
    # Preventing overfitting since the dataset is relatively small 
    # model.add(Dropout(0.1))
    
    # Output Layer
    model.add(Dense(
        units=vocab_size, 
        activation='softmax'))

    model.compile(
        loss='categorical_crossentropy', 
        optimizer='adam', 
        metrics=['accuracy'])
    
    return model

model = create_model(max_sequence_len, vocab_size, X)
model.summary()
```

# Model fit and metrics

```python
model.fit(
    x=X, 
    y=y, 
    epochs=1000, 
    verbose=1)
```

```python
loss, accuracy =  model.evaluate(x=X, y=y)
print(f'Loss: {loss}\nAccuracy: {accuracy}')
```

# Generating product names

```python
def product_name_generator(seed_text, model, num_gen_words, max_sequence_length):
    generated_output = seed_text

        
    for i in range(num_gen_words):
        
        # sequencing the given word(s)
        seed_token_list = tokenizer.texts_to_sequences([generated_output])[0]
        seed_token_list_padded = pad_sequences([seed_token_list], maxlen=max_sequence_length, padding='pre')

        predicted_index = np.argmax(model.predict(seed_token_list_padded), axis=-1)[0]

        predicted_word = tokenizer.index_word[predicted_index]

        generated_output += ' ' + predicted_word
        
    return generated_output    
```
